# TWAP Comparison

```python
from micro_price_trading import TWAP
from micro_price_trading.utils import CI
```

## Setup

```python
quick_twap = TWAP(trade_interval=3, simulations_file='simulations.npy')
medium_twap = TWAP(trade_interval=9, simulations_file='simulations.npy')
slow_twap = TWAP(trade_interval=30, simulations_file='simulations.npy')
```

## Random TWAP

```python
r11, r12 = quick_twap.random_twap()
r51, r52 = medium_twap.random_twap()
r101, r102 = slow_twap.random_twap()
```

```python
(CI(r11), CI(r12))
```

```python
(CI(r51), CI(r52))
```

```python
(CI(r101), CI(r102))
```
